Terör olaylarının değerlerinin mutluluk puanlarından etkilenmediğini, ancak etkisinin 2016’da arttığını gördük. 2018, 2019 ve 2020 verileri için Terörizm verileri olmadığından yalnızca kendi verileriyle analizi yapıldı.
